# Import Library

```python
import pandas as pd
```

```python
df_januari = pd.read_csv('dataset_pdam-bulan-januari.csv')
df_februari = pd.read_csv('dataset_pdam-bulan-februari.csv')
df_maret = pd.read_csv('dataset_pdam-bulan-maret.csv')
df_april = pd.read_csv('dataset_pdam-bulan-april.csv')
df_mei = pd.read_csv('dataset_pdam-bulan-mei.csv')
```

# Merge Data

```python
# List semua DataFrame yang ingin digabungkan
pdam_list = [df_januari, df_februari, df_maret, df_april, df_mei]

# Menggabungkan semua DataFrame menjadi satu
df = pd.concat(pdam_list, ignore_index=True)
df
```

# Drop unused Column

```python
df = df.drop(columns=["commentUrl", "error", "errorDescription", "inputUrl", "id", "likesCount", "media/first_party_cdn_proxied_images/fixed_height/url", 
                      "media/id", "media/images/fixed_height/url", "owner/fbid_v2", "owner/full_name", "owner/id", 
                      "owner/is_mentionable", "owner/is_private", "owner/is_verified", "owner/latest_reel_media", 
                      "owner/profile_pic_id", "owner/profile_pic_url", "owner/username", "ownerProfilePicUrl", 
                      "replies", "repliesCount", "requestErrorMessages/0", "timestamp", "url"])
df = df.rename(columns={"text" : "comment_text"})
df = df[["postUrl", "comment_text", "ownerUsername"]]
df
```

# Add Post Date

```python
date_januari = pd.read_csv('Date Data/january.csv')
date_februari = pd.read_csv('Date Data/february.csv')
date_maret = pd.read_csv('Date Data/march.csv')
date_april = pd.read_csv('Date Data/april.csv')
date_mei = pd.read_csv('Date Data/may.csv')
```

```python
# 2. Gabungkan semua data tanggal menjadi satu DataFrame referensi
df_dates = pd.concat([date_januari, date_februari, date_maret, date_april, date_mei], ignore_index=True)

# 3. Pastikan kolom 'date' bertipe datetime agar bisa diurutkan dengan benar
# Parameter dayfirst=True digunakan karena format Anda adalah DD/MM/YYYY
df_dates['date'] = pd.to_datetime(df_dates['date'], dayfirst=True)

# 4. Cocokkan (Merge) variabel df dengan df_dates berdasarkan 'post_url'
# Kita gunakan how='left' agar data asli di df tidak hilang jika ada post_url yang tidak memiliki tanggal
df = pd.merge(df, df_dates, on='postUrl', how='left')

# Jika di variabel 'df' Anda nama kolomnya 'postUrl' (tanpa underscore), gunakan baris ini sebagai gantinya:
# df = pd.merge(df, df_dates, left_on='postUrl', right_on='post_url', how='left').drop(columns=['post_url'])

# 5. Urutkan data berdasarkan kolom 'date' dari yang terlama ke terbaru
df = df.sort_values(by='date', ascending=True).reset_index(drop=True)

# 6. (Opsional) Kembalikan format tampilan tanggal menjadi DD/MM/YYYY jika dibutuhkan
# Catatan: Setelah diubah ke string format ini, kolom tidak bisa di-sort secara kronologis lagi, 
# jadi lakukan ini di tahap paling akhir setelah sorting selesai.
df['date'] = df['date'].dt.strftime('%d/%m/%Y')
df.to_csv('dataset_pdam_merged.csv', index=False)
df
```
