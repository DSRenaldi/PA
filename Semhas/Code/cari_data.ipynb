```python
import pandas as pd

df_classified = pd.read_csv("Output_V2/train_data.csv")
df_missclassified = pd.read_csv("Output_V2/trial_missclassified_multi_aspect_data.csv")

# Pastikan tipe data sama
df_classified['comment_id'] = df_classified['comment_id'].astype(str)
df_missclassified['comment_id'] = df_missclassified['comment_id'].astype(str)

# Hitung jumlah kemunculan tiap comment_id di classified_sentiment.csv
id_counts = df_classified['comment_id'].value_counts()

# Ambil comment_id yang hanya muncul 1 kali (tidak terduplikasi/tidak tersegmentasi)
unique_ids = id_counts[id_counts == 1].index

# Filter missclassified yang comment_id-nya termasuk unique_ids
single_missclassified = df_missclassified[df_missclassified['comment_id'].isin(unique_ids)] 

# Simpan hasil ke CSV baru
single_missclassified.to_csv("single_missclassified_sentiment.csv", index=False)
```
